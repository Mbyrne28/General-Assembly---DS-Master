```python
import numpy as np
import pandas as pd
import sqlite3

# Open connection to database
db_connection = sqlite3.connect('Chinook_Sqlite.sqlite')
```

## Example

Select the first 10 customers.

```python
pd.read_sql(
    '''SELECT *
       FROM customer
       LIMIT 10''', con=db_connection)
```
